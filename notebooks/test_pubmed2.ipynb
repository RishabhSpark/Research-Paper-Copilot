```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

from app.retriever.pubmed.pubmed_client import fetch_pmc_full_paper
from app.retriever.pubmed.pubmed_parser import extract_text_from_pmc_xml
from app.retriever.pubmed.pubmed_retriever import fetch_pmc_papers
```

```python
def pipeline(query: str, top_k: int = 5):
    list_of_pmc_papers = fetch_pmc_papers(query=query, top_k=top_k)

    for paper in list_of_pmc_papers:
        pmcid = paper
        xml_text = fetch_pmc_full_paper(pmcid=pmcid)
        full_text = extract_text_from_pmc_xml(xml_text=xml_text)
        print(f"PMC ID: {pmcid}")
        print(full_text)
        print("---------------------------------")
```

```python
pipeline("pfizer", 5)
```
